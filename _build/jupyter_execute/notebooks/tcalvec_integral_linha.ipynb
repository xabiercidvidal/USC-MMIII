
### Integral de liña

####  Revisión da parametrización da liña.

 Vimos anteriormente que podemos parametrizar unha liña, ${\bf c}(t)$, dun espazo $\mathbb{R}^n$, con $n>1$, como función vectorial definida nun intervalo $ [t_a, t_b]$ de $\mathbb{R}$.



*Exemplo* : parametrización dun círculo de raio $r$:

 $$
 t \in [0, 2 \pi) \to \left( r \cos \omega t , r \sin \omega t \right)
 $$

 con $\omega = 1$

 *Exemplo* : parametrización dunha hélice de $n$ voltas, de raio $r$:

 $$
 t \in [0, 2 n \pi] \to \left( r \cos \omega t, r \sin \omega t, v_z t \right)
 $$

 con $v_z = 1, \; \omega = 1$.


```python
r, w, vz, n = 1., 1., 1., 4
trange = (0., 2*n*np.pi, 100)
xfun = lambda t : r  * np.cos( w * t)
yfun = lambda t : r  * np.sin( w * t)
zfun = lambda t : vz * t
gf.line3d(xfun, yfun, zfun, trange);
```


Calculabamos a lonxitude do arco, $s$, dunha recta, mediante a integral nunha dimensión dada por:

$$
s = \int_{{\bf c}} \, \mathrm{d}s = \int_{t_a}^{t_b} | \dot{{\bf c}}(t) | \mathrm{d}t
$$

onde 

$$
{\bf \dot{c}}(t) = \left( \frac{\mathrm{d} x_1 (t)}{\mathrm{d}t}, \dots, 
\frac{\mathrm{d}x_n(t)}{\mathrm{d}t}\right)
$$

 é a derivada de ${\bf c}(t)$.

 e

 $$
 \mathrm{d}s = | {\bf \dot{c}}(t)| \, \mathrm{d}t
 $$

 é o arco diferencial. Ten en conta que é unha cantidade escalar.

 En física, se $t$ é o tempo, ${\bf c}(t)$ é a traxectoria dun móbil, ${\bf \dot{c}}(t)$ é a velocidade e $ \mathrm {d}{ s} = |{\bf \dot{c}}(t)| \, \mathrm{d}t$, a distancia percorrida diferencial nun diferencial de tempo $\mathrm{d}t$.



#### Integral dunha función escalar ao longo dunha recta

 Considera que a cerca está situada nunha liña, ${\bf c}(t)$, e ten unha altura, $f(x, y)$, en cada punto desa liña, $f({\bf c} ( t)) $. A área da cerca, $\mathcal{A}$, será a integral da altura, $f(x, y)$ veces a diferencial do arco, $\mathrm{d}s = |{\bf \dot {c}}| \, \mathrm{d}t$, ao longo da liña.

$$
\mathcal{A} = \int_{{\bf c}} f(x, y) \, \mathrm{d} s = \int_{t_a}^{t_b} f({\bf c}(t)) \, |{\bf \dot{c}}(t)| \mathrm{d}t 
$$

 que é simplemente unha integral dunha dimensión en $t$.



*Exercicio* : Supoñamos que o valado está nunha semi-circunferencia de raio unitario, entre os ángulos $\phi \in (-\frac{\pi}{4}, \,\frac{3\pi}{4})$ tal que a súa altura responde á función $f(x, y) = x + y$.


```python
trange = (-np.pi/4, 3*np.pi/4, 60)
cx = lambda t   : np.cos(t)
cy = lambda t   : np.sin(t)
fc = lambda x, y: x + y
gf.int_fscalar_line(fc, cx, cy, trange)
```


A semi-circunferencia do raio unitario pódese parametrizar con:

 $$
 {\bf c}(t) = \left( \cos t, \sin t \right), \;\; t \in (-\frac{\pi}{4}, \,\frac{3\pi}{4})
 $$

 O diferencial da lonxitude do arco:

 $$
 {\bf \dot{c}}(t) = \left( -\sin t, \cos t \right) \to \mathrm{d} s = |{\bf \dot{c}}(t)| \mathrm{d}t = \sqrt{\sin^2 t + \cos^2 t} \; \mathrm{d}t = \mathrm{d}t
 $$

 A integral ao longo do círculo,

$$
\int_{{\bf c}} (x + y) \, \mathrm{d} s = \int_{-\frac{\pi}{4}}^{\frac{3\pi}{4}} (\sin t + \cos t) \, \mathrm{d}t  = -\cos t + \sin t \Big|_{-\frac{\pi}{4}}^{\frac{3\pi}{4}} = 2\sqrt{2}
$$


#### Integral dunha función vectorial ao longo dunha recta

 Definimos a integral dunha función vectorial, ${\bf F}({\bf x})$, ao longo dunha liña, ${\bf c}(t)$, con $t \in [t_a, t_b ]$ , como:

$$
\int_{{\bf c}} {\bf F} ({\bf x}) \, \mathrm{d} {\bf s} = \int_{t_a}^{t_b} {\bf F}({\bf c}(t)) \, {\bf {\dot c}}(t) \, \mathrm{d} t
$$

 onde ${\bf F}({\bf c}(t))$ é a función valorada ao longo da recta e $\mathrm{d} {\bf s} = {\bf \dot{c}} (t) \, \mathrm{d}t$ é a diferencial do arco. Observa que agora $\mathrm{d} {\bf s}$ é un diferencial vectorial e que integramos o produto escalar entre ambos os vectores, ${\bf F} \, \mathrm{d}{\bf s}$.



*Exercicio:* Sexa a función vectorial ${\bf E}(x, y) = (0, E_0)$.

 Calcula a integral ao longo dunha recta horizontal, unha recta vertical e na diagonal.


```python
xrange = (-1.5, 1.5, 11)
trange = (0, 1., 40)
Ex = lambda x, y : 0 + 0.*x
Ey = lambda x, y : 1.+ 0.*x
cx = lambda t    : 0.*t
cy = lambda t    : 1.*t
gf.line2d(cx, cy, trange); 
cx = lambda t    : 1.*t
cy = lambda t    : 0.*t
gf.line2d(cx, cy, trange,color="red",newfig=False); 
cx = lambda t    : 1/np.sqrt(2)*t
cy = lambda t    : 1/np.sqrt(2)*t
gf.line2d(cx, cy, trange,color="green",newfig=False); 
gf.quiver2d(Ex, Ey, xrange, xrange);
```

As traxectorias pedidas son, para as liñas horizontal, vertical e diagonal de lonxitude 1:

 $$
 {\bf c}(t) = (t,0), \; t \in (0, 1)
 $$
 
 $$
 {\bf c}(t) = (0,t), \; t \in (0, 1)
 $$
 
 $$
 {\bf c}(t) = \frac{1}{\sqrt{2}} (t,t), \; t \in (0, 1)
 $$
 
Polo tanto, o produto, con ${\bf F}(x, y) = (0, E_0)$.:
 
$$
{\bf F}({\bf c}(t)) \, \dot{{\bf c}}(t) = 0
$$

$$
{\bf F}({\bf c}(t)) \, \dot{{\bf c}}(t) = E_0
$$

$$
{\bf F}({\bf c}(t)) \, \dot{{\bf c}}(t) = \frac{E_0}{\sqrt{2}}
$$

respectivamente. As integrais serían, 0, $E_0$ e $\frac{E_0}{\sqrt{2}}$, respectivamente.


*Cuestión* : Calcula a integral da función ${\bf F}(x,y) = (x, y)$ ao longo do círculo de raio unitario.


```python
xrange = (-1.5, 1.5, 21)
trange = (0, 2.*np.pi, 40)
Ex = lambda x, y : x
Ey = lambda x, y : y
cx = lambda t    : np.cos(t)
cy = lambda t    : np.sin(t)
gf.line2d(cx, cy, trange); 
gf.quiver2d(Ex, Ey, xrange, xrange);
plt.gca().set_aspect('equal', adjustable='box')
```

Na literatura é normal atopar a expresión:
    
$$
\mathrm{d} {\bf s} = (\mathrm{d}x, \mathrm{d} y) \to {\bf {\dot c}} \, \mathrm{d} t = 
(\dot{x}, \dot{y}) \, \mathrm{d}t
$$

E polo tanto:

$$
{\bf F} \, \mathrm{d}{\bf s} = (F_x, F_y) \, (\mathrm{d}x, \mathrm{d} y) =F_x \mathrm{d}x + F_y \mathrm{d}y \to (F_x \dot{x} + F_y \dot{y}) \, \mathrm{d}t
$$

    


*Observa* : A seguinte figura mostra o campo ${\bf F}(x, y) = (y, 0)$ e a elipse con eixes $a, b$, que neste caso percorremos no sentido horario (comezando por exemplo en ( -a, 0)). Faite unha idea de como é o produto escalar entre o campo ${\bf F}$ e $\mathrm{d}{\bf s}$ ao longo da elipse. Onde é máis grande? Onde é nulo?


```python
a, b = 2., 1.
xrange = (-2.0, 2.0, 13)
trange = (0, 2.*np.pi, 41)
Ex = lambda x, y :      y * 1./2.
Ey = lambda x, y :      x * 0.
cx = lambda t    : a * np.cos(t)
cy = lambda t    : b * np.sin(t)
gf.line2d(cx, cy, trange); 
gf.quiver2d(Ex, Ey, xrange, xrange);
```


*Exercicio* : Agora integra ${\bf F}(x, y) = (-y, 0)$ na elipse dos exes $a, b$, percorridos en sentido antihorario.



Podemos parametrizar a elipse con:

 $$
 {\bf c}(t) = \left(a \cos t, b \sin t \right), \; t \in [0, 2 \pi)
 $$

 a súa derivada é:

 $$
 \dot{{\bf c}}(t) = \left(-a \sin t, b \cos t \right)
 $$

 O campo ${\bf F}(x, y) = (-y, 0)$ valorado na elipse é:

 $$
 {\bf F}({\bf c}(t)) = (-b \sin t, 0)
 $$

 Polo tanto, o produto:
 
 $$
 {\bf F}({\bf c}(t)) \, \dot{{\bf c}}(t) = ab \, \sin^2 t = ab \frac{( 1 - \cos 2 t )}{2}
 $$

 Entón:

 $$
 \int_{{\bf c}} {\bf F} \, \mathrm{d}{\bf s} = \int_0^{2\pi} \frac{ab}{2}(1-\cos 2t) \mathrm{d}t = \frac{ab}{2} \left(t - \frac{\sin 2t}{2} \right) \Big|_0^{2\pi}= ab \pi
 $$


*Exercicio* : Integra ${\bf F}(x, y) = (-y, x)/2$ na elipse de eixos $a, b$, percorrida en sentido antihorario.



Podemos parametrizar a elipse con:

 $$
 {\bf c}(t) = \left(a \cos t, b \sin t \right), \; t \in [0, 2 \pi)
 $$

 a súa derivada é:

 $$
 \dot{{\bf c}}(t) = \left(-a \sin t, b \cos t \right)
 $$

 O campo ${\bf F}(x, y) = (-y, x)/2$ valorado na elipse é:

 $$
 {\bf F}({\bf c}(t)) = (-b \sin t, a \cos t)/2
 $$

 Polo tanto, o produto:

 $$
 {\bf F}({\bf c}(t)) \, \dot{{\bf c}}(t) = ab \, (\sin^2 t +\cos^2 t)/2 = ab /2
 $$

 Entón:

 $$
 \int_{{\bf c}} {\bf F} \, \mathrm{d}{\bf s} = \int_0^{2\pi} \frac{ab}{2} \mathrm{d}t = \frac{ab}{2} t \big|_0^{2\pi}= ab \pi
 $$
