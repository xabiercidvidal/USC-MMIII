
### Teorema de Green

####  Relación entre a integral de liña dun campo vectorial e a integral de superficie.

 Considera unha rexión, $R$, no plano $(x, y)$ definida de tipo I:

 $$
 { a \le x \le b, \; y_1(x) \le y \le y_2(x)}
 $$

 onde $y_1(x)$ e $y_2(x)$ son as funcións que limitan a parte inferior e superior da rexión.

 Ten en conta que podemos parametrizar a liña do límite inferior da rexión con:

 $$
 x \in [a, b], \; {\bf c_1}(x) = \left(x, y_1(x) \right)
 $$

 e a superior con:

 $$
 x \in [a, b], \; {\bf c_2}(x) = \left( x, y_2(x) \right)
 $$



*Observa* : A seguinte figura mostra o campo ${\bf F}(x, y) = (y, 0)$ e a elipse dos eixes $a=2$ en $x$ e $b=1$ en $ y$ . Lembra que a área da elipse é $ab \pi$.


```python
a, b = 2., 1.
xrange = (-2.0, 2.0, 13)
trange = (0, 2.*np.pi, 41)
Ex = lambda x, y :     y * 1.
Ey = lambda x, y :     x * 0.
cx = lambda t    : a * np.cos(t)
cy = lambda t    : b * np.sin(t)
gf.line2d(cx, cy, trange); 
gf.quiver2d(Ex, Ey, xrange, xrange);
```


*Observa* : A seguinte figura mostra como podemos definir a rexión da elipse tipo I, e parametrizar as dúas curvas que determinan o límite da rexión.


```python
trange = (0., np.pi, 60)
gf.line2d(cx, cy, trange, color = 'blue');
plt.text(0.,  0.75, "$y_2(x)$", fontsize = 14);
trange = (np.pi, 2.*np.pi, 60)
gf.line2d(cx, cy, trange, newfig = False, color = 'red');
plt.text(0.,  0.75, "$y_2(x)$", color = 'blue', fontsize = 14);
plt.text(0., -0.75, "$y_1(x)$", color = 'red' , fontsize = 14);
plt.text(.9, 0, "$a$", color = 'black' , fontsize = 16);
plt.plot([1.], [0], marker='o', color = 'black' );
plt.text(-.95, 0, "$b$", color = 'black' , fontsize = 16);
plt.plot([-1.], [0], marker='o', color = 'black' );
```

Calculamos agora a área da rexión:
    
$$
\int_R \mathrm{d}x \mathrm{d} y = \int_a^b \left[ \int_{y_1(x)}^{y_2(x)} \mathrm{dy} 
\right]\mathrm{d}x 
$$

$$
= \int_a^b \left(y_2(x) - y_1(x) \right) \, \mathrm{d} x = \int_a^b y_2(x) \,\mathrm{d}x - \int_a^b y_1(x) \,\mathrm{d}x = \int_a^b y_2(x) \,\mathrm{d}x + \int_b^a y_1(x) \,\mathrm{d}x
$$


Calculamos a integral da función vectorial ${\bf F}(x, y) = (y, 0)$ ao longo de ${\bf c}_1(x)$. 

Dado que ${\bf c}_1 = (x, y_1(x))$, o elemento vectorial diferencial do arco é:

$$
\mathrm{d} {\bf s} = {\bf {\dot c}}(x) \,\mathrm{d} x = \left(1, \dot{y}_1(x) \right) \mathrm{d} x
$$

E o produto escalar:

$$
{\bf F}({\bf x}) \, \mathrm{d} {\bf s} \to (y_1(x), 0) \, \left(1, \dot{y}_1(x) \right) \, \mathrm{d}x = y_1(x) \, \mathrm{d}x
$$

Polo tanto:

$$
\int_{{\bf c}_1} {\bf F}({\bf x}) \, \mathrm{d} {\bf s} 
= \int_{{\bf c}_1} y \, \mathrm{d}x
= \int_a^b y_1(x) \mathrm{d}x
$$

Que corresponde á área comprendida entre a liña $y_1(x)$ (ou ${\bf c}_1$) e o eixe $y=0$, no intervalo $[a, b]$ de $x$.

Así mesmo, a integral da función vectorial, ${\bf F}(x, y) = (y, 0)$, ao longo da liña superior, ${\bf c}_2(x)$ que dá o bordo superior de a rexión:

$$
\int_{{\bf c}_2} {\bf F}({\bf x}) \, \mathrm{d} {\bf s} 
= \int_{{\bf c}_2} y \, \mathrm{d}x
= \int_a^b y_2(x) \, \mathrm{d} x
$$

Se enlazamos a integral recta da función vectorial $(y, 0)$ ao longo de ${\bf c}_2(x)$ e despois ${\bf c}_1(x)$ (en sentido oposto), temos unha integral ao longo dunha curva pechada, ${\bf c} $, no sentido horario, que corresponde á área da rexión, $R$, encerrada por ${\bf c}$.

$$
\oint_{c} (y, 0) \, \mathrm{d} {\bf s} = \oint_{\bf c} y \, \mathrm{d}x  =  \int_{{\bf c}_2} y \, \mathrm{d}x + \int_{{\bf - c}_1} y \, \mathrm{d}x = \int_{{\bf c}_2} y \, \mathrm{d}x - \int_{{\bf c}_1} y \, \mathrm{d}x
$$
$$
= \int_a^b \ y_2(x) \mathrm{d} x - \int_a^b y_1(x)  \mathrm{d} x = \int_{R} \mathrm{d}x \mathrm{d} y
$$

onde

$$
\int_{{\bf c}_1} y = \int_a^b y_1(x)~~~{\mathrm e} ~~~\int_{{\bf -c}_1} y = \int_b^a y_1(x) = - \int_{{\bf c}_1} y
$$

*Cuestión:* Verifica que a integral da función vectorial ${\bf F}(x, y) = (0, x)$ ao longo da fronteira, ${\bf c}(y)$, en sentido anti-horario, dunha rexión, $R$, de $(x, y)$ parametrizada de tipo II, é igual á área de $R$.

$$
\oint_{{\bf c}} (0, x) \, \mathrm{d} {\bf s} 
= \oint_{{\bf c}} x \, \mathrm{d}y
= \int_{R} \mathrm{d}x \mathrm{d}y
$$

#### Teorema de Green

 **Teorema** : A integral do campo vectorial, ${\bf F}(x, y) = \left( F_x(x, y), \; F_y(x, y)\right)$, con $F_x(x, y) ), F_y(x, y)$ con primeiras derivadas continuas nunha rexión, $R$, ao longo dunha liña fronteira, ${\bf c}$ da rexión $R$, atravesada en sentido antihorario é :

$$
\oint_{{\bf c}} {\bf F}(x, y) \, \mathrm{d} {\bf s} = 
\oint_{{\bf c}} F_x(x, y) \, \mathrm{d}x + F_y(x, y) \, \mathrm{d}y \\ = 
\int_{R} \left( \frac{\partial F_y(x, y)}{\partial x} - \frac{\partial F_x(x, y)}{\partial y} \right) \mathrm{d}x \mathrm{d}y
$$


**Lema** : Sexa unha función vectorial, ${\bf F}(x, y) = \left( F_x(x, y), 0 \right)$ con $F_x(x, y)$ con primeiras derivadas continuas nun rexión, $R$; a integral de ${\bf F} $ ao longo dunha liña fronteira, ${\bf c}$, de $R$, percorrida no sentido antihorario é:

$$
\oint_{{\bf c}} {\bf F}(x,  y) \, \mathrm{d} {\bf s} = 
\oint_{{\bf c}} F_x(x, y) \, \mathrm{d}x  =
\int_{R}  - \frac{\partial F_x(x, y)}{\partial y} \mathrm{d}x \mathrm{d}y
$$

Consideremos que definimos a rexión tipo I, con dúas liñas fronteira, parametrizadas en función de $x$ no intervalo $[a, b]$.

$$
\{ a \le x \le b, \; y_1(x) \le y \le y_2(x)\}
$$

con:

$$
{\bf c}_1(x) = \left( x, y_1(x) \right), \; x \in [a, b]
$$

$$
{\bf c}_2(x) = \left( x, y_2(x) \right), \; x \in [a, b]
$$



A integral (fíxate no cambio de signo):

$$
\int_{R}  \frac{\partial F_x(x, y)}{\partial y}  \mathrm{d}x \mathrm{d}y = 
\int_a^b \left[ \int_{y_1(x)}^{y_2(x)}   \frac{\partial F_x(x, y)}{\partial y}  \mathrm{d}y \right]  \mathrm{d}x
$$

$$
 =  \int_a^b  F_x(x, y) \Big|_{y_1(x)}^{y_2(x)} \mathrm{d}x
 =   \int_a^b F_x(x, y_2(x)) \, \mathrm{d}x - \int_a^b F_x(x, y_1(x)) \, \mathrm{d}x
$$


Por outra banda, a integral na recta ${\bf c}_2(x)$ da función vectorial ${\bf F}(x, y ) = \left(F_x(x, y), 0 \right ) $ é:


$$
\int_{{\bf c}_2} {\bf F}(x, y) \, \mathrm{d}{\bf s} 
= \int_{{\bf c}_2} \left(F_x(x, y), 0\right) \, \mathrm{d}{\bf s}  
$$

Como no caso anterior:

$$
\mathrm{d} {\bf s} = {\bf {\dot c}_2}(x) \, \mathrm{d}x = \left(1, \dot{y}_2(x) \right) \mathrm{d}x
$$

E polo tanto:

$$
\int_{{\bf c}_2} \left(F_x(x, y), 0 \right) \,\mathrm{d} {\bf s} 
= \int_a^b \left( F_x(x, y_2(x)), 0 \right) \, \left( 1, \dot{y}_2(x) \right) \, \mathrm{d} x  \\
= \int_a^b F_x(x, y_2(x)) \, \mathrm{d}x
$$

que corresponde ao primeiro termo da integral anterior.

Se entón calculamos a integral do campo ${\bf F}(x, y) = \left( F_x(x, y), 0 \right)$ no sentido horario da curva pechada, ${\bf c }$, que é o límite de $R$, e que corresponde integrar en ${\bf c}_2$ e despois en ${\bf c}_1$ (en sentido contrario), é dicir, no sentido das agullas do reloxo, obtemos:

$$
\oint_{{\bf c}} {\bf F}(x, y) \, \mathrm{d}{\bf s} = 
\int_{{\bf c}_2} {\bf F}(x, y) \, \mathrm{d}{\bf s} - \int_{{\bf c}_1} {\bf F}(x, y) \, \mathrm{d}{\bf s} =
$$
$$
 \int_a^b F_x(x, y_2(x)) \, \mathrm{d}x - \int_a^b F_x(x, y_1(x)) \, \mathrm{d}x =   \\ \int_R \frac{\partial F_x(x, y)}{\partial y} \mathrm{d}x \mathrm{d}y
$$

Se consideramos o sentido contrario ás agullas do reloxo, implica un cambio de signo no lado dereito da igualdade:

$$
\oint_{{\bf c}} \left(F_x(x, y), 0 \right) \mathrm{d} {\bf s} = \int_{R} -\frac{\partial F_x(x, y)}{\partial y} \mathrm{d}x \mathrm{d}y
$$

 co que verificamos o lema.

*Cuestión* : verificar agora este segundo lema:

 **Lema** : A integral da función vectorial ${\bf F}(x, y) = \left( 0, F_y(x, y) \right)$ ao longo da curva ${\bf c}$, que limita unha rexión $R$, no sentido antihorario é:

$$
\oint_{{\bf c}} {\bf F}(x, y) \, \mathrm{d}{\bf s} = \oint_{{\bf c}} F_y(x, y) \, \mathrm{d} y = \int \frac{\partial F_y(x, y)}{\partial x} \mathrm{d}x\mathrm{d}y
$$

 Coa demostración de ambos lemas, compróbase o teorema de Green.

**Corolario:** a integral no sentido antihorario da función vectorial ${\bf F}(x, y) = \left( - y, x \right) /2$ ao longo da liña límite ${\bf c }$ dunha rexión $R $ corresponde á súa área:

 O termo que aparece no teorema de Green é:

$$
\left( \frac{\partial F_y(x, y)}{\partial x} - \frac{\partial F_x (x, y)}{\partial y} \right) = \frac{1}{2} + \frac{1}{2} = 1
$$

Polo tanto:

$$
\oint_{{\bf c}} \frac{1}{2}(-y, x)\, \mathrm{d} {\bf s} = 
\oint_{{\bf c}} -\frac{y}{2} \, \mathrm{d}x + \frac{x}{2} \mathrm{d}y = 
\int_{R} \mathrm{d}x \mathrm{d}y
$$


### Teorema da diverxencia 2D.

 Consideremos a integral dunha función vectorial bidimensional, ${\bf F}({\bf x})$, *a través* dunha curva pechada, ${\bf c}(t)$.

 Dado o elemento vectorial diferencial de arco $\mathrm{d} {\bf s} = (\mathrm{d}x, \mathrm{d}y)$, en lugar de calcular o produtor escalar do campo ao longo da traxectoria, $ { \bf F}(x, y) \, \mathrm{d}{\bf s}$, calculámolo a través del, normal a el, ${\bf F}(x, y) \ , \mathrm{d }{\bf n}$, onde $\mathrm{d}{\bf n}$ é normal a $\mathrm{d}{\bf s}$, pero co mesmo módulo, onde definimos $\mathrm {d}{\bf n} = ( \mathrm{d}y, -\mathrm{d}x)$.

 Isto é:

$$
\int_{{\bf c}} {\bf F}(x, y) \, \mathrm{d}{\bf n}  = \int_{{\bf c}} F_x(x, y) \, \mathrm{d}y - F_y(x, y) \,\mathrm{d}x
$$


O que podemos calcular a partir da parametrización da curva:

 ${\bf c}(t) = (x(t), y(t))$, con $t \in [t_0, t_1]$.

 $$
 \mathrm{d}{\bf n} = (\mathrm{d}y, -\mathrm{d}x) \to (\dot{y}(t), - \dot{x}(t)) \, \mathrm{d}t
 $$

 Polo tanto:

$$
\int_{{\bf c}} {\bf F}(x, y) \, \mathrm{d}{\bf n}  = \int_{t_0}^{t_1} \left( F_x(x(t), y(t)), \, F_y(x(t), y(t)) \right)  \, \left( \dot{y}(t), - \dot{x}(t) \right) \mathrm{d}t
$$



Tomemos agora a integral nunha curva pechada de ${\bf F}(x, y) \, \mathrm{d}{\bf n}$, con $\mathrm{d}{\bf n} = (\mathrm{d}y, - \mathrm{d}x)$

$$
\oint_{{\bf c}} {\bf F}(x, y) \, \mathrm{d}{\bf n}  = \oint_{{\bf c}} F_x(x, y) \, \mathrm{d}y - F_y(x, y) \,\mathrm{d}x
$$

Se chamamos, por visibilidade, $G_x = - F_y, \; G_y = F_x$, e aplicamos o teorema de Green:

$$
\oint_{{\bf c}} {\bf F}(x, y) \, \mathrm{d}{\bf n} =  \oint_{{\bf c}} G_x \mathrm{d} x + G_y \mathrm{d}y =  \int_{R} \left( \frac{\partial G_y}{\partial x} - \frac{\partial G_x}{\partial y}\right) \mathrm{d}x\mathrm{d}y
$$

Se substituímos de novo ${\bf G}(x, y) = (G_x, G_y) = (-F_y, F_x)$, temos:

$$
\oint_{{\bf c}} {\bf F}(x, y) \, \mathrm{d}{\bf n}
= \int_{R} \left( \frac{\partial F_x}{\partial x} + \frac{\partial F_y}{\partial y}\right) \mathrm{d}x\mathrm{d}y
$$

 que se denomina **teorema da diverxencia** .

*Exercicio:* Verifica o teorema da diverxencia 2D co campo vectorial $\mathbf{F} = (x, y)$ nunha circunferencia de radio $r$ centrada na orixe.

### Paso 1: Calculamos a integral de liña

A parametrización da circunferencia de radio $r$ é ${\mathbf c}(t) = r( \cos t,\sin t)$, logo $\mathrm{d} {\mathbf n} = (\mathrm{d}y,-\mathrm{d}x) = (\dot{y},-\dot{x})\mathrm{d}t = r( \cos t,\sin t) \mathrm{d}t$ .

Entón $\mathbf{F} \cdot \mathrm{d} {\mathbf n} = r( \cos t,\sin t) \cdot r( \cos t,\sin t) \mathrm{d}t = r^2 (\cos t^2 +\sin t^2) \mathrm{d}t = r^2 \mathrm{d}t$. Finalmente

$$
\oint_{{\bf c}} {\bf F}(x, y) \, \mathrm{d}{\bf n} = \int_0^{2\pi} r^2 \mathrm{d}t = 2\pi r^2
$$

### Paso 2: Calculamos a integral de área
Temos $\frac{\partial F_x}{\partial x} + \frac{\partial F_y}{\partial y} = 1 + 1 = 2 $. Logo

$$ \int_{R} \left( \frac{\partial F_x}{\partial x} + \frac{\partial F_y}{\partial y}\right) \mathrm{d}x\mathrm{d}y = \int_R 2  \mathrm{d}x\mathrm{d}y $$

Dado que a área da circunferencia é $\pi r^2$, o resultado sería

$$ \int_{R} \left( \frac{\partial F_x}{\partial x} + \frac{\partial F_y}{\partial y}\right) \mathrm{d}x\mathrm{d}y = 2 \pi r^2 $$


1) Calculamos

$$
\frac{\partial F_y}{\partial x} - \frac{\partial F_x}{\partial y} \to 2x -1
$$

Entón:

$$
\int_R \left( \frac{\partial F_y}{\partial x} - \frac{\partial F_x}{\partial y} \right) \mathrm{d}x \mathrm{d}y
\to  \int_0^1 \left[ \int_{x}^{x^2} (2x -1) \mathrm{d}y \right] \mathrm{d}x
$$

$$
= \int_0^1 (2x-1) \, y \big|_x^{x^2} \mathrm{d}x = \int_0^1 (2x-1) (x-x^2) \mathrm{d}x 
$$

$$
= \int_0^1 (-2x^3 + 3x^2 -x) \mathrm{d}x = \left(-\frac{x^4}{2} + x^3 - \frac{x}{2} \right) \Big|_0^1 = 0 
$$




2) Parametrizamos a liña do bordo inferior ${\bf c}_1(x)$ e o bordo superior ${\bf c}_2(x)$:

$$
{\bf c}_1(x) = (x, x^2), \; \dot{{\bf c}}_1(x) = (1, 2x), \; x \in [0, 1]

$$
{\bf c}_2(x) = (x, x), \; \dot{{\bf c}}_2(x) = (1, 1), \; x \in [0, 1]

$$
\int_{{\bf c}_1} {\bf F} \, \mathrm{d}{\bf s} \to \int_0^1 (x^2, x^2) \, (1, 2x) \, \mathrm{d}x 
= \int_0^1 x^2 + 2x^3 \mathrm{d}x \\
= \left(\frac{x^3}{3} + \frac{x^4}{2} \right) \Big|_0^1 
= \frac{5}{6}
$$

2ii) A integral do campo ao longo de ${\bf c}_2$ é:

$$
\int_{{\bf c}_2} {\bf F} \, \mathrm{d}{\bf s} \to \int_0^1 (x, x^2) \, (1, 1) \, \mathrm{d}x 
= \int_0^1 x + x^2 \mathrm{d}x \\
= \left(\frac{x^2}{2} + \frac{x^3}{3} \right) \Big|_0^1 
= \frac{5}{6}
$$

2iii) A integral do campo ao longo de ${\bf c}_1$ e despois ${\bf c}_2$ (en sentido contrario), de tal xeito que atravesemos o límite, ${\bf c}$ , no sentido antihorario é:

$$
\oint_{{\bf c}} {\bf F} \, \mathrm{d} {\bf s} = \int_{{\bf c}_1} {\bf F} \, \mathrm{d} {\bf s} - \int_{{\bf c}_2} {\bf F} \, \mathrm{d} {\bf s} = \frac{5}{6} - \frac{5}{6} = 0
$$
