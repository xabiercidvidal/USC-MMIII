### Funcións escalares

Observa a seguinte figura:

* Viches algunha función en Física semellante á que estamos debuxando?

* Recoñeces o que parece?

```python
def charge_potential():
    rs   = np.linspace(0.1,     10, 60)
    phis = np.linspace(0., 2*np.pi, 60)
    rms, pms = np.meshgrid(rs, phis)
    # compute the electromagnetic potential
    k = 1.; q = 1. # GeV2
    vms =  k*q / rms # potencial  
    vms[vms > 1] = 1. # truncate the figure
    xms = rms * np.cos(pms)
    yms = rms * np.sin(pms)
    fig = plt.figure(); ax = plt.axes(projection='3d')
    sf = ax.plot_surface(xms, yms, vms, cmap='jet');
    plt.xlabel('$x$'); plt.ylabel('$y$'); ax.set_zlabel('$V(x, y)$')
    plt.title('$V(x, y) = 1/ \sqrt{x^2 + y^2}$')
    fig.colorbar(sf); plt.tight_layout();
    return xms, yms, vms
```

```python
xms, yms, vms = charge_potential()
```

É o potencial electromagnético dunha carga no plano

$$
V(x, y) = k \frac{q}{\sqrt{x^2 + y^2}}
$$

Só que tomamos por simplicidade $k = 1$ e $q=1$.

Tamén truncamos o gráfico para velo mellor.

A función anterior é unha **función escalar**.

Unha función escalar, $f({\bf x})$, asigna un valor real, a un punto, ${\bf x}$, dun dominio, $\mathcal{D}$, nun espazo real de dimensión $n$, $\mathbb{R}^n$, con $n>1$.

$$
{\bf x} \in \mathcal{D} \subset \mathbb{R}^n \mapsto f({\bf x}) \in \mathbb{R}
$$

O potencial eléctrico anterior é unha función escalar de $\mathbb{R}^2$, onde o dominio é todo $\mathbb{R}^2$ menos a orixe, $(0, 0)$, onde a función non está definida.

#### Gráfico dunha función

A figura anterior é a **gráfica** do potencial eléctrico dunha carga.

Sexa $f(x, y)$ unha función escalar co dominio $\mathcal{D}$ de $\mathbb{R}^2$, a gráfica da función son os puntos de $\mathbb{R}^3 $ dados por:

$$
\{(x, \, y, \, f(x, y)), \; \forall \;(x, \, y) \, \in \, \mathcal{D}\}
$$


No caso anterior, a gráfica da función $f(x, y)$ son os puntos $(x, y, z = f(x, y))$.

Os humanos somos algo limitados, só podemos "ver" os gráficos de funcións escalares bidimensionais!

#### Conxuntos de nivel

Se observas, o valor do potencial eléctrico é o mesmo para todos os puntos dun círculo centrado na orixe. Dicimos que son puntos equipotenciais.

Doutra maneira, se calculamos cales son os puntos para os cales a función vale un determinado valor $c$, é dicir $f(x, y) = c$, obtemos a ecuación dunha circunferencia:

$$
f(x, y) = c \;\; \rightarrow \frac{1}{\sqrt{x^2 + y^2}} = c\;\; \rightarrow x^2 + y^2 = (1/c)^2
$$


O conxunto de nivel son todos os puntos, ${\bf x} = {x_1, \dots }$ do dominio inicial, $\mathcal{D}$, nun espazo real de dimensión $n$, $\mathbb{R } ^n$, con $n>1$, para os cales o valor da función é o mesmo valor, $c$.
Isto é:

$$
\{  \forall \, {\bf x} \, \in \mathcal{D} \subset \mathbb{R}^n \; | \; f({\bf x}) = c \}
$$

De feito, podemos extraer unha función a partir dos seus conxuntos de nivel.

Observa a seguinte figura. Debuxamos a mesma función, o potencial eléctrico, pero utilizando distintos niveis. Imaxinas a súa gráfica en tres dimensións a partir dos conxuntos de nivel?

Os conxuntos de nivel teñen unha vantaxe adicional, se nos movemos dentro dun deles a función non cambia de valor!

```python
fig = plt.figure()
ax = fig.gca()
sf = ax.contour(xms, yms, vms, 50, cmap='jet');
plt.xlabel('$x$'); plt.ylabel('$y$');  
plt.title('$V(x, y) = 1/ \sqrt{x^2 + y^2}$')
fig.colorbar(sf); plt.axis('equal'); plt.tight_layout();
```

*Exercicio*: Debuxa a gráfica do potencial dun dipolo eléctrico, centrado na orixe, para o cal as cargas, $q$, están separadas por unha distancia $d$. Considera por simplicidade $k=1, q = \pm 1$, $d = 2$.

```python
def dipole_potential():
    dsize = 2.
    xs = np.linspace(-dsize, dsize, 100)
    ys = np.linspace(-dsize, dsize, 100)
    xms, yms = np.meshgrid(xs, ys)

    def Vfield(xs, ys, x0, y0, q0, k = 1.):
        xx, yy = xs - x0, ys - y0
        rr = np.sqrt(xx*xx + yy* yy)
        return k*q0/rr

    V_neg = Vfield(xms, yms, -1., 0., -1.) # field of - charge at (-1, 0)
    V_pos = Vfield(xms, yms, +1., 0., +1.) # field of + charge at (+1, 0)
    V     = V_neg + V_pos
    vmax  = 10. # truncate the potential
    V[V > vmax] = vmax; V[ V < -vmax] = -vmax

    fig = plt.figure(); ax = plt.axes(projection='3d')
    sf = ax.plot_surface(xms, yms, V, cmap = 'jet'); fig.colorbar(sf);
    plt.xlabel('$x$'); plt.ylabel('$y$'); plt.title('V dipolo'); 
    plt.tight_layout();
    return xms, yms, V
```

```python
xms, yms, vdipolo = dipole_potential()
```

```python
fig = plt.figure();
plt.contour(xms, yms, vdipolo, 300, cmap='hot');
plt.xlabel('$x$'); plt.ylabel('$y$'); plt.title('V dipolo')
fig.colorbar(sf,ax=plt.gca()); plt.axis('equal'); plt.tight_layout();
```

### Funcións vectoriais

Considera a seguinte función que seguramente xa teñas visto antes en Física Xeral.

É o campo eléctrico dunha carga $q$ centrada na orixe $(0, 0, 0)$:

$$
{\bf E} (x, y, z) = k\frac{q}{(x^2 + y^2 + z^2)^{3/2}} \left( x, y, z \right )
$$

Quizais o recoñezades mellor se o reescribimos en coordenadas esféricas, onde $r$ é a distancia á carga e ${\hat r}$, o vector unitario que une a carga ata o punto onde estimamos o campo.

$$
{\bf E} ({\bf r}) = k\frac{q}{r^2} {\hat r}
$$

O campo eléctrico, ${\bf E} (x, y, z)$, é unha **función vectorial**.
    
Unha función vectorial mapea a cada punto, ${\bf x}$, dunha rexión ou dominio, $\mathcal{D}$ dun espazo real de dimensión $n$, con $n \ge 1$, $\mathbb{R}^n$, un punto nun espazo real de dimensión $m$, con $m >1$, $\mathbb{R}^m$.

$$
{\bf x} \in \mathcal{D} \subset \mathbb{R}^n \mapsto \bf{f}(\bf x) \in \mathrm{R}^m
$$

Ten en conta que $n \ge 1$ e $ m >1$.

#### Funcións dos compoñentes

Se observas o campo eléctrico, ${\bf E}({\bf x})$ expresámolo como tres funcións escalares:

$$
E_x ( x, y, z) =  k \frac{q}{(x^2+y^2+z^2)} \; \frac{x}{\sqrt{x^2 + y^2 + z^2}} 
$$
$$
E_y ( x, y, z) =  k \frac{q}{(x^2+y^2+z^2)} \; \frac{y}{\sqrt{x^2 + y^2 + z^2}} 
$$
$$
E_z ( x, y, z) =  k \frac{q}{(x^2+y^2+z^2)} \; \frac{z}{\sqrt{x^2 + y^2 + z^2}} 
$$

Cada unha delas é unha función compoñente.

Chamamos á función compoñente $f_i({\bf x})$, a cada función escalar que representa a $i$-ésima coordenada dunha función vectorial,

$$
{\bf f} ({\bf x}) = \left( f_1 ({\bf x}), \dots, f_m({\bf x}) \, \right).
$$

Ou o que é o mesmo, para dar unha función vectorial abonda con dar as súas funcións compoñentes.

*Nota*: Para indicar valores ou funcións vectoriais usamos letras en negra, mentres que usamos tipos normais para valores escalares ou funcións.

#### Como debuxar un campo vectorial?

En realidade, non podemos debuxar correctamente un campo vectorial.

Porén, podemos 'visualizar' un campo vectorial colocando uns pequenos vectores ou frechas no espazo inicial, que polo seu tamaño, grosor ou cor nos indican en que dirección vai o campo e cal é o seu módulo. Aínda que só podemos facelo en $\mathbb{R}^2$ e *mal* en $\mathbb{R}^3$!

A seguinte figura tenta visualizar o campo eléctrico dunha carga en dúas dimensións.

```python
def charge_electric_field():

    xs = np.linspace(-10., 10., 50)
    ys = np.linspace(-10., 10., 50)
    xms, yms = np.meshgrid(xs, ys)

    rms = np.sqrt(xms*xms + yms*yms)
    k, q = 1., 1.

    Ex =  k * q/(rms*rms) * (xms/rms)
    Ey =  k * q/(rms*rms) * (yms/rms)

    fig = plt.figure(); ax = plt.gca()
    color = 2 * np.log(np.hypot(Ex, Ey))
    st = ax.streamplot(xms, yms, Ex, Ey, color = color, linewidth=1, cmap=plt.cm.inferno,
                  density=2, arrowstyle='->', arrowsize=1.5)
    plt.xlabel('$x$'); plt.ylabel('$y$'); plt.title('$(E_x, E_y)$'); 
    fig.colorbar(st.lines); plt.axis('equal'); plt.tight_layout();
    return xms, yms, Ex, Ey
```

```python
xms, ysm, Ex, Ey = charge_electric_field()
```

O campo eléctrico é radial e vaise debilitando a medida que nos afastamos da carga.

*Exercicio*:

Tenta debuxar o campo eléctrico no plano producido por un dipolo, centrado na orixe, con cargas separadas por unha distancia $d =2$. Considera por simplicidade $k =1, \, q = \pm 1$.

```python
def dipole_electric_field():
    dsize    = 3.
    xs       = np.linspace(-dsize, dsize, 50)
    ys       = np.linspace(-dsize, dsize, 50)
    xms, yms = np.meshgrid(xs, ys)
 
    def Efield(xs, ys, x0, y0, q0, k = 1.):
        xx, yy = (xs - x0), ys - y0
        rr = np.sqrt(xx*xx + yy* yy)
        Ex, Ey = k*q0*xx/rr**3, k*q0*yy/rr**3
        return Ex, Ey

    Ex_neg, Ey_neg = Efield(xms, yms, -1., 0., -1.) 
    Ex_pos, Ey_pos = Efield(xms, yms, +1., 0., +1.) 
    Ex    , Ey     = Ex_neg + Ex_pos, Ey_neg + Ey_pos

    color = 2 * np.log(np.hypot(Ex, Ey))
    fig   = plt.figure(); ax = plt.gca()
    st    = ax.streamplot(xms, yms, Ex, Ey, color = color, 
                          linewidth=1, cmap=plt.cm.inferno,
                          density=2, arrowstyle='->', arrowsize=1.5)
    plt.xlabel('$x$'); plt.ylabel('$y$'); plt.title('$(E_x, E_y)$'); 
    fig.colorbar(st.lines); plt.axis('equal'); plt.tight_layout();
    return xms, yms, Ex, Ey
```

```python
xms, ysm, Ex, Ey = dipole_electric_field()
```

As liñas de campo eléctrico dun dipolo saen da carga positiva e morren na carga negativa.
